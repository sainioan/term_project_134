## Binary classifier - Logistic

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics


df = pd.read_csv('utils/C_npf_train.csv')
Y2 = df['class2']
Y4 = df['class4']
df = df.drop(['class4', 'class2'], axis=1)
x_train, x_test, y_train, y_test = train_test_split(df, Y2, test_size=0.25, random_state=0)
df
```

```python

logReg2 = LogisticRegression(max_iter=500)
logReg2.fit(x_train, y_train)

pred2 = logReg2.predict(x_test)

acc2 = logReg2.score(x_test, y_test)
print(acc2)

conf = metrics.confusion_matrix(y_test, pred2)
print(conf)
```

## Multi-label classifier - Logistic

```python
x_train4, x_test4, y_train4, y_test4 = train_test_split(df, Y4, test_size=0.25, random_state=0)
logReg4 = LogisticRegression(max_iter=500)
logReg4.fit(x_train4, y_train4)

pred4 = logReg4.predict(x_test4)

acc4 = logReg4.score(x_test4, y_test4)
print(acc4)

conf4 = metrics.confusion_matrix(y_test4, pred4)
print(conf4)
```
