```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn 

import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.naive_bayes import BernoulliNB

print("hi")
```

```python
trainData = pd.read_csv("npf_train.csv")
trainData.head(3)
```

```python
class2 = np.array(["event"]*trainData.shape[0],dtype="object")
class2[trainData["class4"]=="nonevent"] = "nonevent"
trainData["class2"] = class2
trainData["class2"] = trainData["class2"].astype("category")

trainData.head(3)
```

```python
trainData = trainData.set_index("date")
trainData["class4"] = trainData["class4"].astype("category")
trainData = trainData.drop(["id", "partlybad"],axis=1)

trainData.head(3)
```

```python
trainData.insert(0, 'class22', trainData['class2'])
trainData = trainData.drop('class2', 1)
trainData = trainData.rename(columns={'class22': 'class2'})

trainData['Result'] = trainData['class2'].map(lambda x: x == "event")
trainData['Result'] = trainData['Result'].astype(int)
trainData.insert(0, 'resultV', trainData['Result'])
trainData = trainData.drop('Result', 1)
trainData = trainData.rename(columns={'resultV': 'result'})
trainData.head(5)
```

```python
X = trainData.iloc[:, 3:-1]
y = trainData.iloc[:, 0]

# best features (30 trees)
#X= X[['NET.mean', 'NOx84.mean', 'NOx168.mean', 'NOx336.mean', 'NOx42.mean', 'RHIRGA42.mean', 'RHIRGA168.mean', 'RHIRGA672.mean', 'RHIRGA336.mean', 'RHIRGA168.std', 'NO42.std', 'PAR.mean', 'NOx84.std', 'SO2168.mean', 'SO2168.std', 'T168.std', 'UV_A.mean', 'NO84.mean', 'NO504.std', 'RHIRGA42.std', 'O3168.std', 'CS.mean', 'H2O336.std']]

# worst features (30 trees)
#X= X[['NOx168.std', 'NO84.std', 'NOx42.std', 'NO672.std', 'NO504.mean', 'NO672.mean', 'NO84.mean', 'UV_B.mean', 'CO2168.std', 'CO2336.std', 'O342.mean', 'CO2504.mean', 'CO2504.std', 'O384.mean', 'NO168.std', 'O3168.mean', 'RHIRGA336.std', 'NET.std', 'SWS.mean', 'O3504.mean', 'SO2168.mean', 'RPAR.std', 'H2O168.mean']]

print(X.shape)
print(y.shape)

#X = X.loc[:,['mean' in i for i in X.columns]]
X.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_test_original = X_test
#print(X_test_original)
```

```python
cs = StandardScaler()
#X_train = X_train.values.reshape(1,-1) 
#X_test = X_test.values.reshape(1,-1)
X_train = cs.fit_transform(X_train)
X_test = cs.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
m = RandomForestClassifier(n_estimators=30, random_state=0)
m.fit(X_train, y_train)
y_pred = m.predict(X_test)
print(X_train.shape)
print(X_test.shape)
```

```python
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print("---")
original_score = m.score(X_test, y_test)
print(original_score)
```
